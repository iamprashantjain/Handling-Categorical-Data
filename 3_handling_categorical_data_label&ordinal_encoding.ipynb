```python
#encoding categorical data
#data is of 2 types:

# 1) numerical : like age, weight, height, price etc
# 2) categorical: sex, class, degree, city, country etc
```

```python
#categorical data is of 2 types:

# 1) Nominal: categorical data which has no order for ex: states, hr, up, mp, kerela.. 
#         you cant say up comes first & mp comes last, it has no order
        
# 2) Ordinal: categorical data which order like grades: first, secod, third etc 
#         or political leader: MLA, CM, PM etc which has order 
```

```python
#how to encode categorical data?
#bcoz ml algo expect numbers only & not strings
```

```python
#for categorical data:

#if X (input col) has ordinal categorical data then use ordinal encodeing
#if y (output col) has categorical data then use label encoding
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
#check what kind of data you have in each columns

#age is numerical
#gender is categorical: nominal
#review is categorical: ordinal
#education is categorical: ordinal
#purchased is categorical: nominal but since output is categorical so apply label encoding
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])

#if i dont pass this hirarchy info then it will decided randomly
# may be UG gets last no. & School gets highest no. which is incorreect
```

```python
#fit on train & transform on train & test both

oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```
